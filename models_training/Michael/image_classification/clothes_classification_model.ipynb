```python
data_path = '/home/ec2-user/SageMaker/datasets/Clothes_Classification_Data'
```

```python
import cv2
import ultralytics
from ultralytics import YOLO

from datetime import datetime
curr_date = datetime.now().date()
print(curr_date)
```

```python
#Initialize clearml task
from clearml import Task, StorageManager, Dataset


task = Task.init(project_name = "exxos_models",
                 task_name = f"clothes_classification_model_{curr_date}")
```

```python
model = YOLO("yolov8n-cls.pt")

args = dict(data=data_path, epochs = 150, 
            imgsz=640, exist_ok = True, verbose = True, device=0)
```

### Train model 

```python
results = model.train(**args)
task.close()
```
